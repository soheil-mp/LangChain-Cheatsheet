## Step 6 - Prompt Engineering

---

<br>

##### Step 1 - Chain of Thought (CoT) Template

```python
# Chain of Thought (CoT) Prompt Template
cot_template = """
You are a knowledgeable AI assistant. Approach questions step-by-step, provide detailed explanations, and use critical thinking.

To answer any question, follow these steps:

1. Analyze the question and identify key components:
   [Your analysis here]

2. Gather relevant information and recall pertinent knowledge:
   [Your gathered information here]

3. Consider different perspectives or approaches:
   [Your consideration of different perspectives here]

4. Evaluate the information and form a conclusion:
   [Your evaluation and conclusion here]

5. Provide a clear and concise answer:
   [Your clear answer here]

6. Explain your reasoning:
   [Your explanation here]

Final Answer: [Provide your final, summarized answer here]
"""
```

<br>

##### Step 2 - Example Selector (Few-Shot Learning)

```python
# Prepare examples
examples = [
    {"input": "What is 5 + 3?", "output": "To solve 5 + 3:\n1. Add the numbers: 5 + 3 = 8\nThe answer is 8."},
    {"input": "Calculate 7 - 2", "output": "To solve 7 - 2:\n1. Subtract 2 from 7: 7 - 2 = 5\nThe answer is 5."},
    {"input": "Multiply 4 and 6", "output": "To multiply 4 and 6:\n1. Multiply the numbers: 4 * 6 = 24\nThe answer is 24."},
    {"input": "Divide 15 by 3", "output": "To divide 15 by 3:\n1. Divide 15 by 3: 15 / 3 = 5\nThe answer is 5."},
    {"input": "What's the area of a rectangle with length 8 and width 5?", "output": "To find the area of a rectangle with length 8 and width 5:\n1. Use the formula: Area = length * width\n2. Calculate: 8 * 5 = 40\nThe area is 40 square units."}
]

# Create a ChatPromptTemplate from a list of message templates
example_prompt = ChatPromptTemplate.from_messages([
    HumanMessagePromptTemplate.from_template("{input}"),
    AIMessage(content="{output}")
])

# Create a FewShotChatMessagePromptTemplate using the example prompt
few_shot_prompt = FewShotChatMessagePromptTemplate(
    example_prompt=example_prompt,
    examples=examples,
    input_variables=["input"]
)
```

<br>

##### Step 3 - Final Prompt

```python
# Final Chat Prompt Template
final_prompt = ChatPromptTemplate.from_messages([

    # CoT prompt
    SystemMessagePromptTemplate.from_template(cot_template),

    # Memory placeholder
    MessagesPlaceholder(variable_name="history"),

    # Few-shot prompt
    few_shot_prompt,

    # Human message
    # HumanMessagePromptTemplate.from_template("I need some advice about {input}")
])
```

```python
# Json Tool
from langchain.tools.json.tool import JsonSpec, JsonGetValueTool
data = {"name": "John", "age": 30, "city": "New York"}
json_spec = JsonSpec(dict_=data)
json_tool = JsonGetValueTool(spec=json_spec)
```

```python
# Function calling tool

# Calculate the time difference in minutes between two dates
def calculate_time_difference(dates: str) -> int:
    date1, date2 = dates.split(',')
    d1 = pd.to_datetime(date1.strip())
    d2 = pd.to_datetime(date2.strip())
    difference = abs((d2 - d1).total_seconds() / 60)
    return int(difference)

# Create a Tool for using the function
func_tool =  Tool(
        name="TimeDifferenceTool",
        func=calculate_time_difference,
        description="Useful for calculating the time difference in minutes between two dates. Input should be two dates in the format 'YYYY-MM-DD HH:MM:SS' seperated by comma."
)
```
